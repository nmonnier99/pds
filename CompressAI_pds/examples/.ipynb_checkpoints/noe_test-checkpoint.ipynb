## Image extraction

```python
def get_kodim_tensor(img_number): 
    img_name = '/kodim{:02d}.png'.format(img_number)
    img = Image.open('./assets'+ img_name).convert('RGB')
    x = transforms.ToTensor()(img).unsqueeze(0).to(device)
    return img, x
```

```python
def get_anchor_image(anchor, quality, img_number): 
    '''
    Inputs : 
        anchor = int (1 or 3) 
        quality = int (1 to 8)
        img_number = int (1 to 24)
    '''
    anchor_path = 'assets/anchor' + str(anchor)
    img_name = '/{}kodim{:02d}.jpeg'.format(quality, img_number)
    img = Image.open(anchor_path+ img_name).convert('RGB')
    return img
```

```python
def get_anchor_dna(anchor, quality, img_number): 
    anchor_path = 'assets/anchor' + str(anchor)
    dna_name = '/{}kodim{:02d}.fasta'.format(quality, img_number)
    fasta_file = open(anchor_path + dna_name)
    dna = fasta_file.read()
    fasta_file.close()
    return dna
```

```python
def get_jpeg_image(img_number, quality):
    img_path = 'assets/'
    img_name = 'kodim{:02d}_decoded{}.png'.format(img_number, quality)
    img = Image.open(img_path+ img_name).convert('RGB')
    return img
```

### Function computing DNA and reconstruction image 

```python
def compute_x_hat(x, quality):
    
    net = bmshj2018_factorized(quality, pretrained=True).eval().to(device)

    #COMPRESS
    with torch.no_grad():
        dictionary = net.get_y_hat_and_medians(x)
        y_hat = dictionary['y_hat']
        medians = dictionary['medians']

    coder = Coder() ## Class defined in coder.py

    y_rounded = y_hat-torch.unsqueeze(medians, dim=0)
    y_rounded = torch.round(y_rounded).int()

    #ENCODE
    dna = coder.encode(y_rounded, quality) 

    #DECODE
    y_rounded_decoded = coder.decode(dna, quality, x.shape)
    y_hat_decoded = y_rounded_decoded + torch.unsqueeze(medians, dim=0)

    #DECOMPRESS
    with torch.no_grad():
        dict_dna = net.get_x_hat(y_hat_decoded)
        out_net_dna = {'x_hat': dict_dna['x_hat']}
        out_net_dna['x_hat'] = out_net_dna['x_hat'].clamp(0, 1)

    x_hat =  out_net_dna['x_hat'].clamp(0, 1)
    return x_hat, dna
```

```python
for i in range(1,2): 
    for q in range(1, 2): 
        img, x = get_kodim_tensor(i)
        x_hat, dna = compute_x_hat(x, q)
        
        path = '/assets'
        img_name = '{}kodim{:02d}_decoded.png'.format(q, i)
        img_path = path + img_name
        imageio.imwrite(img_path, x_hat)
        
        dna_name = '{}kodim{:02d}_dna.fasta'.format(q, i)
        dna_path = path + dna_name
        with open(dna_path, 'w') as f:
            f.write(dna)
```

### Homopolymers

```python
def count_homopolymers(count, homo, dna, i) :
    if i == 0 :
        return count 
    elif dna[i] == dna[i-1] : 
        count_homopolymers(count, homo+1, dna, i-1)
    elif homo >= 2 : 
        count.append(homo)
        count_homopolymers(count, 0, dna, i-1)
    else : 
        count_homopolymers(count, 0, dna, i-1) 
```

### Metrics

```python
def compute_psnr(a, b):
    mse = torch.mean((a - b)**2).item()
    return -10 * math.log10(mse)

def compute_msssim(a, b):
    return ms_ssim(a, b, data_range=1.).item()

def compute_bit_rate(img, dna): 
    h, w = img.size
    return len(dna)/(h*w)

```

### PSNR

```python
def plot_PSNR(img_number): 
    
    '''
    Plots PSNR versus bit rate for a test image and two anchor images.
    '''
    
    PSNR_test = []
    PSNR_a1 = []
    PSNR_a3 = []
    PSNR_C = []
    bit_rates_t = []
    bit_rates_1 = []
    bit_rates_3 = []
    
    
    img = Image.open('./assets'+ '/kodim{:02d}.png'.format(img_number)).convert('RGB')
    x = transforms.ToTensor()(img).unsqueeze(0).to(device)
    
    for q in range (1,8):
        #Test 
        x_hat, dna = compute_x_hat(x, q)
        PSNR_test.append(compute_psnr(x, x_hat))
        br = compute_bit_rate(img, dna)
        bit_rates_t.append(br)
        
        #PSNR anchor1
        imgA1 = get_anchor_image(1, q, img_number)
        imgA1_tensor = transforms.ToTensor()(imgA1).unsqueeze(0).to(device)
        dnaA1 = get_anchor_dna(1, q, img_number)
        PSNR_a1.append(compute_psnr(x, imgA1_tensor))
        bit_rates_1.append(compute_bit_rate(imgA1, dnaA1))
        
        
        #PSNR anchor3
        '''imgA3 = get_anchor_image(3, q, img_number)
        dnaA3 = get_anchor_dna(3, q, img_number)
        imgA3_tensor = transforms.ToTensor()(imgA3).unsqueeze(0).to(device)
        PSNR_a3.append(compute_psnr(x, imgA3_tensor))
        bit_rates_3.append(compute_bit_rate(imgA3, dnaA3))
        
        #JPEG DNA CODEC
        imgC = get_jpeg_image(img_number, q)
        imgC_tensor = transforms.ToTensor()(imgC).unsqueeze(0).to(device)
        PSNR_C.append(compute_psnr(x, imgC_tensor))
        '''
        
        
    # plot results
    plt.plot(bit_rates_t, PSNR_test, label='Test', marker='o', markersize=3)
    plt.plot(bit_rates_1, PSNR_a1, label='Anchor 1', marker='s', markersize=3)
    #plt.plot(bit_rates_3, PSNR_a3, label='Anchor 3', marker='^', markersize=3)
    plt.xlabel('Bit rate (ntpp)')
    plt.ylabel('PSNR (dB)')
    plt.title('PSNR vs. Quality')
    plt.legend()
    plt.show()
```

### MSSSIM

```python
def plot_MSSSIM(img_number): 
    
    '''
    Plots MSSSIM versus bit rate for a test image and two anchor images.
    '''
    
    MSSSIM_test = []
    MSSSIM_a1 = []
    MSSSIM_a3 = []
    bit_rates_t = []
    bit_rates_1 = []
    bit_rates_3 = []
    
    
    img = Image.open('./assets'+ '/kodim{:02d}.png'.format(img_number)).convert('RGB')
    x = transforms.ToTensor()(img).unsqueeze(0).to(device)
    
    for q in range (1,8):
        #Test 
        x_hat, dna = compute_x_hat(x, q)
        MSSSIM_test.append(compute_msssim(x, x_hat))
        br = compute_bit_rate(img, dna)
        bit_rates_t.append(br)
        
        #MSSSIM anchor1
        imgA1 = get_anchor_image(1, q, img_number)
        imgA1_tensor = transforms.ToTensor()(imgA1).unsqueeze(0).to(device)
        dnaA1 = get_anchor_dna(1, q, img_number)
        MSSSIM_a1.append(compute_msssim(x, imgA1_tensor))
        bit_rates_1.append(compute_bit_rate(imgA1, dnaA1))
        
        
        #MSSSIM anchor3
        imgA3 = get_anchor_image(3, q, img_number)
        dnaA3 = get_anchor_dna(3, q, img_number)
        imgA3_tensor = transforms.ToTensor()(imgA3).unsqueeze(0).to(device)
        MSSSIM_a3.append(compute_msssim(x, imgA3_tensor))
        bit_rates_3.append(compute_bit_rate(imgA3, dnaA3))
        
        
    # plot results
    plt.plot(bit_rates_t, MSSSIM_test, label='Test', marker='o', markersize=3)
    plt.plot(bit_rates_1, MSSSIM_a1, label='Anchor 1', marker='s', markersize=3)
    plt.plot(bit_rates_3, MSSSIM_a3, label='Anchor 3', marker='^', markersize=3)
    plt.xlabel('Bit rate (ntpp)')
    plt.ylabel('MSSSIM (dB)')
    plt.title('MSSSIM vs. Quality')
    plt.legend()
    plt.show()
    
```
